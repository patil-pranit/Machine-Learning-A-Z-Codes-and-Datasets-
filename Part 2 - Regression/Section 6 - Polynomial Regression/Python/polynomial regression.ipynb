```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv('Position_Salaries.csv')
X = df.iloc[:, 1:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
reg1 = LinearRegression()
lin_reg1 = reg1.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_feat = PolynomialFeatures(degree=4)
X_ploy = poly_feat.fit_transform(X)
reg2 = LinearRegression()
lin_reg2 = reg2.fit(X_ploy, y)
```

```python
plt.scatter(X, y, c='red')
plt.plot(X, lin_reg1.predict(X), color = 'blue')
plt.show()
```

```python
plt.scatter(X, y, c='red')
plt.plot(X, lin_reg2.predict(poly_feat.fit_transform(X)), color = 'blue')
plt.show()
```

```python
print(lin_reg1.intercept_)
print(lin_reg1.coef_)
```

```python
print(lin_reg2.intercept_)
print(lin_reg2.coef_)
```
